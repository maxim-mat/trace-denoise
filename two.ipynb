```python
import pickle as pkl
from torch.utils.data import DataLoader
import pandas as pd
import numpy as np
import warnings

warnings.filterwarnings('ignore')
```

```python
with open("data/pickles/50salads_target.pkl", "rb") as f:
    salads_target = pkl.load(f)

with open("data/pickles/50salads_softmax.pkl", "rb") as f:
    salads_predict = pkl.load(f)
```

```python
from sklearn.preprocessing import OneHotEncoder

def one_hot_encode(l):
    encoder = OneHotEncoder(sparse_output=False)
    return encoder.fit_transform(np.array(l).reshape(-1, 1))
```

```python
one_hot_salads = [one_hot_encode(x) for x in salads_target]
```

```python
with open("data/pickles/50_salads_one_hot.pkl", "wb") as f:
    pkl.dump(one_hot_salads, f)
```
